```python
import random
import numpy as np

nomes = [
    "IronThroneBot", "DragonForge", "WinterDroid", "StormBreaker", "ShadowViper",
    "IceWarden", "FireDrake", "NightSentinel", "BloodRavenX", "SteelDirewolf",
    "DracarysX", "FrostClaw", "TitanLannister", "ValyrianCore", "StormbornX",
    "DreadfortBot", "GhostProtocol", "RavenEye", "ObsidianKnight", "Kingslayer9000",
    "TheHoundBot", "FrozenBastion", "BattleOfIce", "DorneRacer", "MechaViper",
    "WildfireEngine", "BranSentinel", "ShadowOfHarrenhal", "IronFleetX", "TheWallGuardian"
]

def gerar_participante():
    nome = random.choice(nomes)
    vel = random.uniform(5, 15)
    return {"nome": nome, "vel": vel, "pos": 0, "tempo": 0}

def alterar_desempenho(robo):
    chance = random.random()
    if chance < 0.1:
        ajuste = random.uniform(0.1, 0.5)
        robo["vel"] *= (1 - ajuste)
        robo["vel"] = max(robo["vel"], 5)
        print(f"{robo['nome']} penalizado! Velocidade: {robo['vel']:.2f} m/s")
    elif chance < 0.2:
        ajuste = random.uniform(0.1, 0.3)
        robo["vel"] *= (1 + ajuste)
        print(f"{robo['nome']} ganhou bônus! Velocidade: {robo['vel']:.2f} m/s")

def simular_evento(quantidade):
    quantidade = min(quantidade, 10)
    competidores = [gerar_participante() for _ in range(quantidade)]
    print("\nParticipantes:")
    for i, robo in enumerate(competidores):
        print(f"{i+1}. {robo['nome']} (Velocidade inicial: {robo['vel']:.1f} m/s)")

    print("\nCorrida iniciada...")
    distancia = 1000
    while True:
        for robo in competidores:
            alterar_desempenho(robo)
            nova_pos = robo["pos"] + robo["vel"]
            robo["pos"] = min(nova_pos, distancia)
            robo["tempo"] += 1
        if all(robo["pos"] >= distancia for robo in competidores):
            break

    competidores.sort(key=lambda x: x["tempo"])
    print("\nCorrida finalizada!\n")
    print("=== Resultados ===")
    for i, robo in enumerate(competidores):
        print(f"{i+1}º lugar: {robo['nome']} (Tempo: {robo['tempo']:.1f} s)")

    vel_media = distancia / competidores[0]["tempo"]
    diferenca = competidores[-1]["tempo"] - competidores[0]["tempo"]
    print("\n=== Estatísticas ===")
    print(f"- Velocidade média do vencedor: {vel_media:.2f} m/s")
    print(f"- Diferença de tempo entre o 1º e o último: {diferenca:.1f} s")

print("=== Corrida de Robôs ===")
num = int(input("Quantos robôs participarão? (Máx: 10) "))
simular_evento(num)
```
